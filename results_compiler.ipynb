```python
def get_acc_and_var(results, save_path=None):
    results_acc_over_experiment = np.mean(results, axis=0).round(2)
    results_std_over_experiment = np.std(results, axis=0).round(3)
    
    results_acc_over_subject = np.mean(results_acc_over_experiment, axis=0).round(2)
    results_std_over_subject = np.mean(results_std_over_experiment, axis=0).round(3)
    
    print(results_acc_over_experiment)
    print(results_acc_over_subject)
    print()
    print(results_std_over_experiment)
    print(results_std_over_subject)
    
    results_acc = np.concatenate((results_acc_over_experiment, np.atleast_2d(results_acc_over_subject)), axis=0)
    results_std = np.concatenate((results_std_over_experiment, np.atleast_2d(results_std_over_subject)), axis=0)
    
    return results_acc, results_std
```

```python
LDA_2_acc, LDA_2_std = get_acc_and_var(LDA_2)
LDA_CSP_2_acc, LDA_CSP_2_std = get_acc_and_var(LDA_CSP_2)
LDA_SACSP_2_acc, LDA_SACSP_2_std = get_acc_and_var(LDA_SACSP_2)
LDA_4_acc, LDA_4_std = get_acc_and_var(LDA_4)
LDA_CSP_4_acc, LDA_CSP_4_std = get_acc_and_var(LDA_CSP_4)
LDA_SACSP_4_acc, LDA_SACSP_4_std = get_acc_and_var(LDA_SACSP_4)
```
